```python
dn=pd.read_csv('pregnancy.csv') #pregnancy csv file(7283 rows Ã— 3 columns) 
dn.dropna()
dn=dn.drop(["Unnamed: 0"],axis=1)

df2=pd.read_csv('Demographic.csv')#The required Demographic dataset(59842 rows Ã— 5 columns).
df2=df2.drop(['Unnamed: 0'],axis=1)

df=df.drop(['Unnamed: 0'],axis=1)#df containing the BMI_Index of the entries.
```

```python
l=[2.0]
dn=dn[dn.URXPREG.isin(l) == False]
# So now we are cleaning the data further with avaliable,
# 1 represents positive and 3 means Not done.
```

```python
merge = pd.merge(df, dn, on="SEQN",how="left")
# We merge the pregnancy and BMI dataset.
merge1 = pd.merge(df2,merge,on='SEQN',how='left')#df2-demographic data
merge2 = pd.merge(merge1,db,on='SEQN',how='left')#db is the whole dataset.
# The dataset -(796264 rows Ã— 91 columns)
```

```python
a=merge2.index[merge2['URXPREG'] == 1.0].tolist()
merge2=merge2.drop(merge2.index[a])
merge2=merge2.drop(['URXPREG'],axis=1)
```

```python
merge2.to_csv('forKMeans.csv')
```
